```python
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
from cartopy.util import add_cyclic_point
```

```python
file = xr.open_dataset('pr_Amon_CESM1-CAM5_amip_r1i1p1_197901-200512.nc')
```

```python
lat = file['lat'][:]
lon = file['lon'][:]
var = file['pr'][:,:,:]
```

```python
clim = var.groupby('time.month').mean('time')
```

```python
cjan, clon = add_cyclic_point(clim[0].values, coord=lon)
cfev, clon = add_cyclic_point(clim[1].values, coord=lon)
cmar, clon = add_cyclic_point(clim[2].values, coord=lon)
cabr, clon = add_cyclic_point(clim[3].values, coord=lon)
cmai, clon = add_cyclic_point(clim[4].values, coord=lon)
cjun, clon = add_cyclic_point(clim[5].values, coord=lon)
cjul, clon = add_cyclic_point(clim[6].values, coord=lon)
cago, clon = add_cyclic_point(clim[7].values, coord=lon)
cset, clon = add_cyclic_point(clim[8].values, coord=lon)
cout, clon = add_cyclic_point(clim[9].values, coord=lon)
cnov, clon = add_cyclic_point(clim[10].values, coord=lon)
cdez, clon = add_cyclic_point(clim[11].values, coord=lon)
```

```python
meses = [cjan, cfev, cmar, cabr, cmai, cjun, cjul, cago, cset, cout, cnov, cdez]
```

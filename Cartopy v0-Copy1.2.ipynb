```python
import netCDF4 as nc
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
from cartopy.util import add_cyclic_point
```

```python
# file = nc.Dataset('sst.mnmean.v4.nc')

file = xr.open_dataset('dadosiri/NovIC_nmme_precip_anom_stdanom.nc')
# sresa1b_ncar_ccsm3-example.nc sst.mnmean.v4.nc
```

```python
file
```

```python
# file

lat = file['lat'][:]
lon = file['lon'][:]
var = file['precip'][:,:,:]
```

```python
cvar, clon = add_cyclic_point(var[0,:,:].values, coord=lon, axis=1)
cvar2, clon = add_cyclic_point(var.values, coord=lon, axis=1)
```

```python
print(len(clon))
print(var.shape)
```

```python
fig, ax = plt.subplots(figsize=(15,10),nrows = 3, ncols = 2, subplot_kw={'projection': ccrs.PlateCarree()}, sharex = False)

for i in range(3):
    for j in range(2):
        
        contour = ax[i, j].contourf(clon, lat, cvar, levels = np.arange(0,20,2),
              transform=ccrs.PlateCarree(),cmap = plt.get_cmap('Blues'), extend='both')
        
        ax[i, j].add_feature(cfeature.BORDERS,zorder=2,edgecolor='k')
        ax[i, j].add_feature(cfeature.COASTLINE)
        
```
